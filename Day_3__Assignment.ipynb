## 1. Given a string consisting of whitespace and groups of matched parentheses, write a Python program to split it into groups of perfectly matched parentheses without any whitespace.

image.png


```python
# Write Your Code here
def split_paranthesis(string):
    list_ = []
    string.replace(' ', '')
    for i in string:
      if i != " ":
        list_.append(i)
        s=''
        new_str = s.join(list_)
    
    return(new_str)
```

```python
a = '( ()) ((()()())) (()) ()'
print(split_paranthesis(a))
```

## 2. Write a Python program to find the strings in a given list, starting with a given prefix.

image.png

```python
# Write Your Code here
def find_string(prefix,list):
  op = []
  for i in list:
    if i[:len(prefix)]==prefix:
      op.append(i)
  return(op)
```

## 7. What is multicollinearity, and how do you treat it and when is it not a problem?

Write your answer here.

Multicollinearity occurs when independent variables in a regression models are correlated to each other. Multicollinearity maximizes the errors in a model hence model cannot predict a nearest prediction output compared to the actual one. 
Multicollinearity can be detected using variance inflation factor (VIF). The collinearity of independent variables can also be checked using correlation heatmap. Based on the heat map we can find out the mostly correlated independent variables.
We can treat multicollinearity by:

*   Removing some highly correlated independent variables
*   linearly combine the independent variables by adding them together

Multicollinearity can be neglected if the dataset contains variables which are not effective in predicting the dependent variables






## 8. VIF(Variation Inflation Factor),Weight of Evidence & Information Value. Why and when to use?

Write your answer here.

**VIF (Variance Inflation Factor)** plays an important role in selecting the independent variables of dataset before feeding it to the machine learning model.
In best practices, if VIF is approximate to 1 , we can say thatvariables are not correlated to each other while having a value between 1 to 5, we can say that variables are moderately correlated to each other while more than 5 VIF might represent highly correlated variables.

**Weight of Evidence(WoE)** can be stated as the factor which tells us the weightage of any independent variables category to predict the Dependent variables. It is used for the categorical data as regression models do not take categorical inputs.

**Information Value** helps us to tell the feature importance of any independent variable to predict the dependent variable.
