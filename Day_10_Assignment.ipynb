## 1. Write a Python program to find the closest palindrome from a given string.

Input:  cat

Output:cac

Input:  madan

Output:madam

Input: radivider

Output:radividar

Input: madan

Output:madam

Input: abc

Output:aba

Input: racecbr

Output:racecar


```python
# Write Your code here.
def palindrome(string):
  palindrome_list=[]
  for i,j in enumerate(string):
    if string[::]!= string[::-1]:
      c=string[-1*(i+1)]
      d = string[i]
      if c!=d:
        b = string.replace(c,d)
        return b
    else:
      return f"The given string {string} is already a palindrome"
```

```python
a = "cat"
b = "madan"
c = "radivider"
d = "abc"
e = "racecbr"
f = "madam"

print(palindrome(a))
print(palindrome(b))
print(palindrome(c))
print(palindrome(d))
print(palindrome(e))
print(palindrome(f))
```

## 2. Write a Python program to find an integer exponent x such that a^x = n.

Input:
a =  2 : n =  1024

Output:
10

Input:
a =  3 : n =  81

Output:
4

Input:
a =  3 : n =  1290070078170102666248196035845070394933441741644993085810116441344597492642263849

Output:
170

image.png

```python
# Write Your code here.
def find_power(a,n):
  lst =  [x for x in range(int(n/a)) if a**x == n]
  return lst[0]
```

```python
print(find_power(2,1024))
print(find_power(3,81))
```

## 4. From the tables write a SQL query to find those students who achieved 100 percent marks in every subject for all the year. Return subject ID, subject name, students for all year.

**Table Name1: exam_test**

**Table Name2: subject_test**

**Expected Output:**

image.png

```python
# Write Your code here.
mysql('''
SELECT  s.subject_id, p.subject_name, 
SUM(s.no_of_student) 'Students for all year'  
FROM exam_test as s 
JOIN subject_test as p 
ON s.subject_id = p.subject_id
GROUP BY s.subject_id
''')
```
