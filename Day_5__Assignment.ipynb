## 1. Write a Python program to create a list whose ith element is the maximum of the first i elements from a input list.

```python
# Write Your code here.
def find_max(input_list):
  output_list = [max(input_list[0:i]) for i in range(1,len(input_list)+1) ]
  return print(f"Input: \n {input_list} \n Output: \n {output_list}")
```

```python
list_1 = [0, -1, 3, 8, 5, 9, 8, 14, 2, 4, 3, -10, 10, 17, 41, 22, -4, -4, -15, 0]
find_max(list_1)
```

## 2. Write a Python program to find the coordinates of a triangle with the given side lengths.

image.png

image.png

image.png

```python
# Write Your code here.
def find_coordinates(sides):
  
  a,b,c = sides                #Length of sides
  s = (a+b+c)/2                #semiperimeter
  A = (s*(s-a)*(s-b)*(s-c))**0.5  #Heron's Formula for area
  h = 2*A /a                   # h=height as we know Area = 1/2 * base * height

  #As we have height now we can calculate the distance between origin and projection oh height on base using pythagoras theorem
  p = ((c**2)-(h**2))**0.5
  return print(f"Input: \n {sides} \n Output: \n {[[0,0],[a,0],[p,h]]}")
```

```python
sides_1 = [3, 4, 5]
print(find_coordinates(sides_1))
sides_2 = [5, 6, 7]
print(find_coordinates(sides_2))
```
