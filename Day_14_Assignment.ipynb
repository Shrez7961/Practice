## 1. Write a Python program to find a substring in a given string contains a vowel between two consonants.

From vocabulary.com:
A consonant is a speech sound that is not a vowel. It also refers to letters of the alphabet that represent those sounds: Z, B, T, G, and H are all consonants. Consonants are all the non-vowel sounds, or their corresponding letters: A, E, I, O, U and sometimes Y are not consonants.

image.png

```python
# Write Your code here.
def find(text):
    cons = "bcdfghjklmnpqrstvwxz"
    vows = "aeiou"
    return next(text[i - 1:i + 2] for i in range(1, len(text) - 1)
                if text[i].lower() in vows and text[i - 1].lower() in cons and text[i + 1].lower() in cons)
```

```python
text_1 = "Hello"
text_2 = "Sandwhich"
print(find(text_1))
print(find(text_2))
```

## 2. Write a Python program to find all 5's in integers less than n that are divisible by 9 or 15.

image.png

```python
# Write Your code here.
def find(n):
    return [[i,j] for i in range(n) for j in range(len(str(i))) if str(i)[j] == '5' and (i%15==0 or i%9==0)]
```

```python
n=50
print(find(n))
n2=65
print(find(n2))
n3=75
print(find(n3))
n4=85
print(find(n4))
```

## I. Which item was purchased just before the customer became a member?

**Expected Output:**

image.png


```python
# Write Your code here.
mysql('''
select a.customer_id, order_id, product_id,
rank() over (partition by  a.customer_id order by s.order_date) as rank
from member as b
inner join sales as a
on a.customer_id = b.customer_id
where order_date < join_date
''')
```

## II. What is the total items and amount spent for each member before they became a member?

**Expected Output:**

image.png

```python
# Write Your code here.
mysql('''
Select S.customer_id,count(S.product_id ) as unique_menu_item ,Sum(M.price) as total_sales
From sales as S
Join menu as M
ON M.product_id = S.product_id
JOIN members Mem
ON Mem.Customer_id = S.customer_id
Where S.order_date < Mem.join_date
Group by S.customer_id
order by total_sales asc
''')
```
