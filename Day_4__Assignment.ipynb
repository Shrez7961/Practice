## 1. An irregular/uneven matrix, or ragged matrix, is a matrix that has a different number of elements in each row. Ragged matrices are not used in linear algebra, since standard matrix transformations cannot be performed on them, but they are useful as arrays in computing.

## Write a Python program to find the indices of all occurrences of target in the uneven matrix.

```python
# Write Your Code here.
def find_index(Matrix, Target):
    return [[i, j] for i, row in enumerate(Matrix) for j, k in enumerate(row) if k == Target]
Matrix = [[1, 3, 2, 32, 19], [19, 2, 48, 19], [], [9, 35, 4], [3, 19]]
Target = 19

print(find_index(Matrix,Target))
```

## 2. Monotonic sequences are sequences, which constantly increase or constantly decrease.
## Write a Python program to determine the direction ('increasing' or 'decreasing') of monotonic sequence numbers.

image.png

```python
# Write Your Code here.
def monotonic(sequence):
  difference=[]
  for i in  range(len(sequence)-1):
    
    diff = sequence[i+1]-sequence[i]
    difference.append(diff)
    a= set(difference)
  if len(a)==1:
    if diff < 0:
      print("This is a monotonically deccreasing sequence")
    elif diff>1:
      print("This is  a monotonically increasing sequence")
    else:
      print("This is not a monotonic sequence")
  else:
    print("This is not a monotonic sequence")
```

```python
series_1 = [10,20,30,40,50,60]
series_2 = [6,5,4,3,2,1]
series_3 = [19,19,5,5,5,5,5]
series_4 = [1,1,1,1,1,1]
print(monotonic(series_1))
print(monotonic(series_2))
print(monotonic(series_3))
print(monotonic(series_4))
```
