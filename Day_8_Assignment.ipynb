## 1. Write a Python program to find the vowels from each of the original texts (y counts as a vowel at the end of the word) from a given list of strings.

image.png

```python
# Write Your code here.

def find_vowel(strs):

  return ["".join(i for i in text if i.lower() in "aeiou") + (text[-1] if text[-1].lower() == "y" else "")
            for text in strs]
```

```python
l = ["w3resource", "Python", "Java", "C++"]
find_vowel(l)
```
